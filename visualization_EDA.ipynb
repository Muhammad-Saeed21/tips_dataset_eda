```python
import pandas as pd        # Pandas for data handling and analysis
import seaborn as sns      # Seaborn for dataset and visualizations
import matplotlib.pyplot as plt   # Matplotlib for plotting graphs
```

```python
df = sns.load_dataset("tips")   # Load the built-in 'tips' dataset from seaborn
df.head()                       # Display the first 5 rows of the dataset
```

```python
df.info()   # Shows column names, number of rows, data types, and null values
```

```python
df.describe()   # Provides basic statistical summary for numeric columns
```

```python
df.isnull().sum()   # Counts missing values in each column
```

```python
df['day'].unique()   # Shows unique days like Thur, Fri, Sat, Sun
```

```python
plt.figure(figsize=(6,4))    # Set figure size
plt.hist(df['total_bill'])   # Create histogram for total bill column
plt.title("Distribution of Total Bill")   # Title of the chart
plt.xlabel("Total Bill")     # X-axis label
plt.ylabel("Frequency")      # Y-axis label
plt.tight_layout()           # Adjust layout to avoid overlapping
plt.show()                   # Display the plot
```

```python
plt.figure(figsize=(6,4))           # Set figure size
plt.scatter(df['total_bill'],        # X-axis: total bill
            df['tip'])               # Y-axis: tip
plt.title("Total Bill vs Tip")       # Title of the chart
plt.xlabel("Total Bill")             # X-axis label
plt.ylabel("Tip")                    # Y-axis label
plt.tight_layout()                   # Adjust layout for neatness
plt.show()                           # Show the plot
```

```python
plt.figure(figsize=(6,4))        # Set figure size
sns.countplot(x='day', data=df)  # Countplot to show number of customers per day
plt.title("Customers per Day")   # Title of the graph
plt.xlabel("Day")                # X-axis label
plt.ylabel("Count")              # Y-axis label
plt.tight_layout()               # Fix graph spacing
plt.show()                       # Display the plot
```
